## 1. Tuples – immutable ordered sequences

```python
# One policy/scheme record: (name, year, budget in lakhs) — immutable
scheme_record = ("Ayushman Bharat", 2023, 6500)
print("Scheme record:", scheme_record)
print("Scheme name:", scheme_record[0])
print("Budget (lakhs):", scheme_record[2])
```

```python
# Unpacking: useful for fixed policy/scheme fields
scheme_name, year, budget_lakhs = scheme_record
print("Unpacked:", scheme_name, year, "Budget:", budget_lakhs, "lakh")

# Tuple from a list (e.g. one row from a program table)
row = tuple(["PM-KISAN", 2024, 60000])  # scheme, year, beneficiaries (approx)
print("Row:", row)
```

## 2. Sets – unique, unordered elements

```python
# Unique beneficiary IDs or unique departments (duplicates removed)
beneficiary_ids = {"B001", "B002", "B001", "B003", "B002"}
print("Unique beneficiaries:", beneficiary_ids)

# From a list – e.g. districts where a scheme was implemented
districts_reported = ["Bargarh", "Sambalpur", "Bargarh", "Sundargarh", "Sambalpur"]
unique_districts = set(districts_reported)
print("Unique districts:", unique_districts)
print("Count:", len(unique_districts))
```

```python
# Set operations: e.g. regions covered by two programs
program_A_regions = {"North", "South", "East"}
program_B_regions = {"South", "East", "West"}
print("All regions (union):", program_A_regions | program_B_regions)
print("Common (intersection):", program_A_regions & program_B_regions)
print("Only in A (difference):", program_A_regions - program_B_regions)
print("South covered by A?", "South" in program_A_regions)
```

## 3. Dictionaries – key–value pairs

```python
# Scheme/policy as key–value: lookup by ID or name
scheme = {
    "code": "AB-PMJAY",
    "name": "Ayushman Bharat PM-JAY",
    "year": 2023,
    "budget_cr": 6500,
    "ministry": "Health"
}
print("Scheme:", scheme)
print("Name:", scheme["name"])
print("Keys:", list(scheme.keys()))
print("Values:", list(scheme.values()))
```

```python
# Iterate over key-value pairs (e.g. for a report line)
for key, value in scheme.items():
    print(f"  {key}: {value}")

# Safe access with .get() — useful when key might be missing
print("Budget (cr):", scheme.get("budget_cr"))
print("Region (if any):", scheme.get("region", "N/A"))
```

## 4. Practical example – unique count and lookup

```python
# Survey/feedback: stakeholder type (with duplicates)
stakeholder_types = ["NGO", "Govt", "NGO", "Private", "Govt", "NGO"]
unique_stakeholders = set(stakeholder_types)
print("Number of unique stakeholder types:", len(unique_stakeholders))
print("Types:", unique_stakeholders)

# Lookup: stakeholder type -> department/contact type (policy context)
stakeholder_info = {
    "NGO": "Civil society",
    "Govt": "Line department",
    "Private": "Corporate/PPP"
}
for s in unique_stakeholders:
    print(s, "->", stakeholder_info.get(s, "Other"))
```

## 5. Practice problems – solutions (all)

Below: solutions for **5.1–5.2** (in-class) and **5.3–5.6** (homework).

### 5.1 Solution — KPI record (tuple)

```python
kpi = ("Enrollment rate (%)", 95, 88)
name, target, achieved = kpi
pct = (achieved / target) * 100 if target else 0
print("Indicator:", name)
print("Achievement %:", round(pct, 1))
```

### 5.2 Solution — Unique scheme codes (set)

```python
scheme_codes = ["AB", "PK", "AB", "MS", "PK", "AB"]
unique_schemes = set(scheme_codes)
print("Number of unique schemes:", len(unique_schemes))
print("Schemes:", unique_schemes)
```

### 5.3 Solution — District code to name (dictionary)

```python
districts = {"D1": "Bargarh", "D2": "Sambalpur", "D3": "Sundargarh"}
codes = ["D1", "D3", "D2", "D4"]
for code in codes:
    print(code, "->", districts.get(code, "Unknown"))
```

### 5.4 Solution — Total amount per scheme

```python
releases = [("AB", 100), ("PK", 50), ("AB", 75), ("PK", 25)]
totals = {}
for code, amount in releases:
    totals[code] = totals.get(code, 0) + amount
print("Totals per scheme:", totals)
```

### 5.5 Solution — Satisfaction survey (set + dict)

```python
responses = ["Satisfied", "Neutral", "Satisfied", "Unsatisfied", "Neutral"]
unique_resp = set(responses)
label = {"Satisfied": "Positive", "Neutral": "Neutral", "Unsatisfied": "Negative"}
for r in unique_resp:
    print(r, "->", label.get(r, "Other"))
```

### 5.6 Solution — Policy milestones (tuples in list)

```python
milestones = [
    ("Phase 1 rollout", "Q2", "Completed"),
    ("Phase 2", "Q3", "Due"),
    ("Review", "Q4", "Pending")
]
for name, quarter, status in milestones:
    if status in ("Due", "Completed"):
        print(f"{name} ({quarter}): {status}")
```
