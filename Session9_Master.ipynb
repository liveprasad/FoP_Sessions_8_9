## 1. Defining and calling a function

```python
def format_scheme_line(name, year):
    """Return a one-line description for a policy/scheme (e.g. for reports)."""
    return f"{name} ({year})"

print(format_scheme_line("PM-JAY", 2023))
print(format_scheme_line("National Education Policy", 2024))
```

## 2. Parameters and return value

```python
def total_budget(allocated, revised):
    """Return total of allocated and revised budget (e.g. in lakhs)."""
    return allocated + revised

print("Total (lakhs):", total_budget(100, 50))
print("Total (lakhs):", total_budget(500, 100))
```

```python
def format_policy_info(name, year):
    """Return a short description string for policy/scheme."""
    return f"Policy: {name} ({year})"

print(format_policy_info("Ayushman Bharat", 2023))
print(format_policy_info(year=2024, name="PM-KISAN"))
```

## 3. Using data structures inside functions

```python
def count_unique(items):
    """Return the number of unique elements (e.g. unique districts, beneficiaries)."""
    return len(set(items))

print(count_unique(["North", "South", "North", "East"]))  # unique regions
print(count_unique(["D1", "D2", "D1"]))  # unique department codes
```

```python
def get_summary(data):
    """Return a short summary dict for a list of numbers (e.g. grant amounts, beneficiaries)."""
    if not data:
        return {"count": 0, "total": 0}
    return {
        "count": len(data),
        "total": sum(data),
        "min": min(data),
        "max": max(data)
    }

# e.g. monthly expenditure in lakhs across districts
expenditure_lakhs = [10, 20, 30, 40]
print(get_summary(expenditure_lakhs))
```

## 4. Practical example – reusable formatter

```python
def format_budget(amount):
    """Format amount in lakhs for display (e.g. scheme budget in reports)."""
    lakhs = amount / 100_000
    return f"₹{lakhs:.1f} lakh"

print(format_budget(500000))   # e.g. district allocation
print(format_budget(1000000))  # e.g. scheme outlay
```

### 6.1 Solution — Format budget in crores

```python
def format_cr(amount):
    cr = amount / 1_00_00_000
    return f"{cr:.1f} crore"

print(format_cr(65000000))
print(format_cr(10000000))
```

### 6.2 Solution — One-line scheme status

```python
def scheme_status(s):
    return f"{s['code']}: {s['budget_cr']} cr ({s['ministry']})"

print(scheme_status({"code": "AB", "budget_cr": 6500, "ministry": "Health"}))
```

### 6.3 Solution — Count above threshold

```python
def count_above(amounts, threshold):
    count = 0
    for a in amounts:
        if a >= threshold:
            count += 1
    return count

print(count_above([50, 75, 30, 80], 50))  # 3
```

### 6.4 Solution — Indicator lookup

```python
def lookup_indicator(indicators, code):
    return indicators.get(code, "Not found")

ind = {"ENR": "Enrollment rate", "LIT": "Literacy"}
print(lookup_indicator(ind, "ENR"))
print(lookup_indicator(ind, "XYZ"))
```

### 6.5 Solution — Report line

```python
def report_line(scheme_name, amount_lakhs):
    return f"Scheme: {scheme_name} | Amount: {amount_lakhs} lakh"

print(report_line("PM-JAY", 100))
```

### 6.6 Solution — Merge totals (two dicts)

```python
def merge_totals(d1, d2):
    result = {}
    for key in d1:
        result[key] = d1[key] + d2.get(key, 0)
    for key in d2:
        if key not in result:
            result[key] = d2[key]
    return result

print(merge_totals({"AB": 100, "PK": 50}, {"AB": 25, "PK": 75}))
```
